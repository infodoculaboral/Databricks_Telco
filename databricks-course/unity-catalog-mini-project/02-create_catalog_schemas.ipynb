####Crear Catalogs y Schemas para el Proyecto
1. Catalog - Telco_dev (sin managed location)
2. Schemas - bronze, silver y gold (con managed location)

```sql
CREATE CATALOG IF NOT EXISTS telco_dev;
```

```sql
USE CATALOG telco_dev;
```

```sql
CREATE SCHEMA IF NOT EXISTS bronze
MANAGED LOCATION "abfss://bronze@sadatabricksext1.dfs.core.windows.net/";
```

```sql
CREATE SCHEMA IF NOT EXISTS silver
MANAGED LOCATION "abfss://silver@sadatabricksext1.dfs.core.windows.net/";
```

```sql
CREATE SCHEMA IF NOT EXISTS gold
MANAGED LOCATION "abfss://gold@sadatabricksext1.dfs.core.windows.net/";
```
